```python
import time
import json
import random
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException, WebDriverException

# Initialize the Chrome driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the webpage
print("Opening the webpage...")
driver.get("https://roomescapeartist.com/find-a-room/")
print("Webpage opened successfully.")

# Switch to the iframe that contains the data
iframe = WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.TAG_NAME, "iframe")))

# Initialize an empty list to hold the data
data_list = []

# Find all table rows
rows = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "table tbody tr")))
total_rows = len(rows)
print(f"Found {total_rows} rows.")

# Define the maximum number of retries
max_retries = 3

# Function to save progress incrementally
def save_progress():
    with open('escape_rooms.json', 'w') as json_file:
        json.dump(data_list, json_file, indent=4)

# Counter to keep track of successful row clicks
success_count = 0

# Iterate through each row and extract data
for index, row in enumerate(rows):
    print(f"Processing row {index + 1}/{total_rows}...")
    retry_count = 0
    while retry_count < max_retries:
        try:
            # Scroll to the row using JavaScript
            driver.execute_script("arguments[0].scrollIntoView(true);", row)
            time.sleep(random.uniform(1, 4))  # Random sleep to mimic human behavior

            # Click the row to open the dialog
            ActionChains(driver).move_to_element(row).click(row).perform()

            # Wait for the dialog to appear
            WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".v-dialog")))

            # Extract data from the dialog
            try:
                name = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".v-card__title"))).text
                address_element = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".mdi-map-marker + span a")))
                address = address_element.text
                address_url = address_element.get_attribute("href")
                website_element = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".mdi-link-variant + span a")))
                website = website_element.get_attribute("href")

                # Append the extracted data to the list
                data_list.append({
                    "name": name,
                    "address": address,
                    "address_url": address_url,
                    "website": website
                })

                # Increment success count
                success_count += 1

                # Close the dialog
                close_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".v-dialog .v-btn__content")))
                close_button.click()

                # Wait for the dialog to close
                time.sleep(random.uniform(2, 7))  # Random sleep to mimic human behavior

                print(f"Successfully processed row {index + 1}.")
                # Break the retry loop if successful
                break

            except Exception as e:
                print(f"Error extracting data from dialog: {e}")
                raise e  # Raise the exception to trigger a retry

        except (TimeoutException, WebDriverException) as e:
            retry_count += 1
            print(f"Error processing row {index + 1} (attempt {retry_count}/{max_retries}): {e}")
            if retry_count == max_retries:
                print("Max retries reached, skipping this row.")
                # Log skipped row
                row_data = {
                    "row_number": index + 1,
                    "status": "skipped",
                    "error_message": str(e)
                }
                data_list.append(row_data)
                # Extract and save data before skipping
                try:
                    name = row.find_element(By.CSS_SELECTOR, ".v-card__title").text
                    address_element = row.find_element(By.CSS_SELECTOR, ".mdi-map-marker + span a")
                    address = address_element.text
                    address_url = address_element.get_attribute("href")
                    website_element = row.find_element(By.CSS_SELECTOR, ".mdi-link-variant + span a")
                    website = website_element.get_attribute("href")
                    row_data["name"] = name
                    row_data["address"] = address
                    row_data["address_url"] = address_url
                    row_data["website"] = website
                except Exception as ex:
                    print(f"Error extracting data from row {index + 1}: {ex}")
                    row_data["data_extraction_error"] = str(ex)

                # Append the row data to the list
                data_list.append(row_data)

            time.sleep(random.uniform(2, 5))  # Random sleep before retrying

    # Save progress after each row
    save_progress()

# Close the browser
driver.quit()

# Final save to ensure all data is written to the file
save_progress()

# Print success count
print(f"Successfully processed {success_count} rows.")
```
